```python
import pandas as pd
import numpy as np
import random
from sentence_transformers import SentenceTransformer, util
import os
# from transformers import Trainer, pipeline, set_seed, AutoTokenizer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer
import nltk
nltk.download('punkt')
import csv
from datasets import load_dataset, load_metric
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.05, random_state=42)
```

```python
train.head()
```

```python
train.head(5)
```

```python
from sklearn.neighbors import NearestNeighbors

# Load your model
model = SentenceTransformer('sentence-transformers/stsb-bert-large')

# Encode the training data
train_embeddings = model.encode(train['text'].tolist())

# Fit the NearestNeighbors model
nn_model = NearestNeighbors(n_neighbors=1, algorithm='brute', metric='cosine')
nn_model.fit(train_embeddings)

# Encode the test data
test_embeddings = model.encode(test['text'].tolist())

# Find the nearest neighbors for each test sentence
distances, indices = nn_model.kneighbors(test_embeddings)

# Get the reframed sentences based on nearest neighbors
flatten_data = [train['rephrase'].iloc[index] for index in indices.flatten()]
```

```python
import nltk
from rouge_score import rouge_scorer

nltk.download('punkt')  # Download the required resource for NLTK

# Calculate BLEU score
def calculate_bleu_score(reference, hypothesis):
    reference = [nltk.word_tokenize(sent) for sent in reference]
    hypothesis = nltk.word_tokenize(hypothesis)
    return nltk.translate.bleu_score.sentence_bleu(reference, hypothesis)

# Calculate ROUGE score
def calculate_rouge_score(reference, hypothesis):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores['rouge1'].fmeasure, scores['rouge2'].fmeasure, scores['rougeL'].fmeasure

# Example usage
reference_texts = test['rephrase'].tolist()

# Calculate scores for each hypothesis in hyp2
bleu_scores = [calculate_bleu_score(reference_texts[i], flatten_data[i]) for i in range(len(flatten_data))]
rouge_scores = [calculate_rouge_score(reference_texts[i], flatten_data[i]) for i in range(len(flatten_data))]

# Print average scores
avg_bleu_score = sum(bleu_scores) / len(bleu_scores)
avg_rouge1_score = sum(score[0] for score in rouge_scores) / len(rouge_scores)
avg_rouge2_score = sum(score[1] for score in rouge_scores) / len(rouge_scores)
avg_rougeL_score = sum(score[2] for score in rouge_scores) / len(rouge_scores)

print(f'Average BLEU Score: {avg_bleu_score}')
print(f'Average ROUGE-1 Score: {avg_rouge1_score}')
print(f'Average ROUGE-2 Score: {avg_rouge2_score}')
print(f'Average ROUGE-L Score: {avg_rougeL_score}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Encode the test data
BART_embeddings = model.encode(df['BART'].tolist())
test_rephrase_embeddings = model.encode(df['test_rephrase'].tolist())

# Calculate cosine similarity between corresponding BART and test_rephrase embeddings
cosine_similarities = [cosine_similarity([bart_embedding], [test_embedding])[0][0]
                       for bart_embedding, test_embedding in zip(BART_embeddings, test_rephrase_embeddings)]

# Create a DataFrame to store the results
result_df_bart = pd.DataFrame({
    'BART_Text': df['BART'],
    'Test_Rephrase_Text': df['test_rephrase'],
    'Cosine_Similarity': cosine_similarities
})

# Display the result DataFrame
print(result_df_bart)
```
